```python
import seaborn as sns
import pandas as pd
```

```python
sns.set(rc={'figure.figsize':[8,8]}, font_scale=1.2)
```

```python
df = pd.read_csv('50ulke.csv')
```

```python
#Filtering data by continents
df_asia=df[df["Continent"]=="Asia"]
df_europe=df[df["Continent"]=="Europe"]
df_africa=df[df["Continent"]=="Africa"]
df_north_america=df[df["Continent"]=="North America"]
df_south_america=df[df["Continent"]=="South America"]
#Growth comparisons among continents (except Australia/Oceania, because it has no variance)
sns.kdeplot(df_asia["Growth"], label = "Asia")
sns.kdeplot(df_europe["Growth"], label = "Europe")
sns.kdeplot(df_africa["Growth"], label = "Africa")
sns.kdeplot(df_north_america["Growth"], label = "North America")
sns.kdeplot(df_south_america["Growth"], label = "South America")
#By observing this graph, it looks like Asia has the lowest and highest growth rates in the world.
#Europe has a positive skew distribution (which means, most of it's countries has low growth rates).
#Africa has a positive skew distribution, also compared to other continents, most of it's countries have low growth rate.
#North America almost has a normal distribution and most of it's growth rates are focused around 2.
#South America also has a normal distribution, most of it's growth rate values are focues around 1.5.
```

```python
#Let's use Europe as an example for relation between growth and national income.
sns.regplot(x=df_europe["Growth"],y=df_europe["National Income"])
#It appears that in Europe, growth rate and national income are inversely proportional.
```

```python
#Further explanation for Europe's growth and national income's correlation.
sns.jointplot(x="Growth", y="National Income", data=df_europe, kind = "scatter")
#Growth has a positive skew distribution.
#National Income has a negative skew distribution.
#Therefore, they are inversely proportional.
```

```python
#Correlations among growth, national income ,and per person in Europe.
sns.pairplot(df_europe)
```

```python
#Comparing continents correlations among growth, national income, and per person.
sns.pairplot(df, hue="Continent")
```

```python
#Comparing growth rates among continents.
sns.boxplot(x="Continent", y="Growth", data= df)
#Asia has the widest range of growth rates in the world
#South America has the narrowest range of growth rates in the world.
#Europe has the highest growth rate in the world.
```

```python
sns.violinplot(x="Continent",y="Growth", data=df)
#North America has the widest Mode, which means most of it's countries have similar growth rates.
#On the contrary, Asia has the biggest variety of growth rates.
```

```python
sns.stripplot(x="Continent",y="Per person",data=df)
#Europe has the widest range of personal income.
#Africa has almost the lowest and narrowest range of personal income.
```

```python
sns.swarmplot(x="Continent", y="National Income", data=df)
#This graph shows that most of Asian countires have similar national income.
```

```python
sns.countplot(df["Continent"])
#This shows that the data source only provided lots of data on Europe and Asia (It has no meaning here).
```

```python
sns.heatmap(df.corr(), annot=True)
#This shows us that there is almost no correlation between national income and growth.
```
